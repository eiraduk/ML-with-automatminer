# ML with automatminer

```python
#loading a dataset
from matminer.datasets import load_dataset
```

```python
df = load_dataset("matbench_steels", data_home=".")
```

```python
df.head(2)
```

```python
#in this dataset we don't have formula, volume ,...
#we need to split the dataframe into 2 parts. one is for training and other is for testing
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df, test_size = 0.2) #spilit dataframe, testing on 20%
```

```python
len(train_df)
```

```python
len(test_df)
```

```python
#create a new variable
target = "yield strength"
test_df = test_df.drop([target], axis =1)
```

```python
train_df.head(3)
```

```python
#our test_df4 will only have composition only
test_df.head(3)
```

```python
#automatminer will now add the features automatically
```

```python
from automatminer import MatPipe
```

```python
pipe = MatPipe.from_preset("express_single") #preset we used is express_single to give general performance. 
#sinngle means means we are not doing the automatic machine learning part, we are going to train as single pipeline
```

```python
#fitting pipe
pipe.fit(train_df, target)
```

```python
#automatminer selected the best featulizer that provided a good performance
#it generated all the features 
#it did data cleaning process by itself
#it did also feature reduction
```

```python
#We can now use pipe object to predict some values

test_df = pipe.predict(test_df)
```

```python
test_df
```

```python
import pprint #for pretty print dictionary
```

```python
summary = pipe.summarize(filename="summary.json")
```

```python
pprint.pprint(summary) #printing summary
```

```python
#create a variable called detail
details = pipe.inspect() #to have more information
```

```python
pprint.pprint(details) #printing summary
```
